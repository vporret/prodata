## Exercise 7

Increasing the size of the training set seems to lead to a higher percentage of correction for the Multi-Layered Perceptron Neural Network, as indicated below

    200:   69.0% Correct
    1000:  75.0% Correct
    5000:  79.9% Correct


```python
import pylab as pl
import numpy as np
import mlp
import cPickle, gzip

# Read the dataset in (code from sheet)
f = gzip.open('mnist.pkl.gz','rb')
tset, vset, teset = cPickle.load(f)
f.close()

nread = [200, 500, 1000]
for k in range(len(nread)):
    #nread = nread[k]
    # Just use the first few images
    train_in = tset[0][:nread[k],:]


    # This is a little bit of work -- 1 of N encoding
    # Make sure you understand how it does it
    train_tgt = np.zeros((nread[k],10))
    for i in range(nread[k]):
        train_tgt[i,tset[1][i]] = 1
    
    test_in = teset[0][:nread[k],:]
    test_tgt = np.zeros((nread[k],10))
    for i in range(nread[k]):
        test_tgt[i,teset[1][i]] = 1
    
    # We will need the validation set
    valid_in = vset[0][:nread[k],:]
    valid_tgt = np.zeros((nread[k],10))
    for i in range(nread[k]):
        valid_tgt[i,vset[1][i]] = 1
    
    for i in [1,2,5,10,20]:
        print ("----- "+str(i))
        net = mlp.mlp(train_in,train_tgt,i,outtype='softmax')
        net.earlystopping(train_in,train_tgt,valid_in,valid_tgt,0.1)
        net.confmat(test_in,test_tgt)
```
